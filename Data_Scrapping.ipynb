# Datascrapping from URA website

```python
#datascraping
import time
from selenium.webdriver.support.ui import Select
```

```python
from selenium import webdriver
# URA Website
website = 'https://www.ura.gov.sg/realEstateIIWeb/transaction/search.action'
# Download Folder
download_folder = 'I:\OneDrive - National University of Singapore\DBA5106 Foundation in BA\Download'
chrome_options = webdriver.ChromeOptions()

prefs = {'download.default_directory' : download_folder}
chrome_options.add_experimental_option('prefs', prefs)
driver = webdriver.Chrome('C:/Python39/Scripts/chromedriver.exe', chrome_options=chrome_options)
driver.get(website)
```

```python
z=0
v=0
max_value = 2970/5+1

for x in range (0,(int(max_value))):
    print('x',x)
    driver.get(website)
    select = Select(driver.find_element_by_id('searchForm_selectedFromPeriodProjectName'))
    select.select_by_value('NOV 2016')
    driver.find_element_by_xpath("//div[@id='property']/div[2]/div/label").click()
    for y in range (0+z,5+z):
        print('v',v)
        if v <2971:
            driver.find_element_by_id('addToProject_{0}'.format(v)).click()
        v+=1
    time.sleep(1)
    driver.find_element_by_id('searchForm_0'.format(v)).click() 
    time.sleep(1)
    # Check if there is Data
    try:
        driver.find_element_by_xpath("//form[@id='SubmitSortForm']/div[2]/div[3]/input").click()
    except:
        print('no data')
    
    z += 5
```
